```python
import csv

# Replace 'existing.csv' with the name of your existing CSV file
existing_csv_file = 'T-HSAB_csv.csv'

# Replace 'new.csv' with the name of the new CSV file with headers
    new_csv_file = 'T-HSAB_csv_h.csv'

# Define the headers you want to add
headers = ["comment", "type"]

# Read the existing CSV file
data = []
with open(existing_csv_file, 'r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
        data.append(row)

# Add the headers to the beginning of the data
data.insert(0, headers)

# Write the data with headers to a new CSV file
with open(new_csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(data)

print(f'Headers have been added to the CSV file. New file: "{new_csv_file}"')
```

```python
 #Text length analysis 
def avg_word(sentence):
    words = sentence.split()
    if len(words) == 0:
        return 0
    return (sum(len(word) for word in words)/len(words))
```

```python
import nltk
nltk.download('stopwords')
stop = set(nltk.corpus.stopwords.words("arabic"))
```

```python
data['word_count'] = data['comment'].apply(lambda x: len(str(x).split(" ")))
data['char_count'] = data['comment'].str.len() ## this also includes spaces
data['avg_char_per_word'] = data['comment'].apply(lambda x: avg_word(x))
data['stopwords'] = data['comment'].apply(lambda x: len([x for x in x.split() if x in stop]))
data = data.sort_values(by='word_count',ascending=[0])
data.head()
```

```python
dataset_stat = data.mean(axis=0)
dataset_stat
```
